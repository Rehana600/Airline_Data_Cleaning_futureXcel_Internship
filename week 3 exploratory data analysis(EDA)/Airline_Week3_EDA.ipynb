# **Load Data**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/cleaned_airline_satisfaction (1) (1).csv")
df.head()
```

# **Overview**

```python

df.info()
df.describe()
```

# **Missing Values**

```python
df.isnull().sum()
```

# **UNIVARIATE**

```python
df['satisfaction'].value_counts()
```

# **Graph**

```python
df['satisfaction'].value_counts().plot(
    kind='bar', color=['green','red']
)
plt.title('Passenger Satisfaction')
plt.show()
```

# **BIVARIATE**

```python
pd.crosstab(df['Class'], df['satisfaction'])
```

# **Graph**

```python
sns.countplot(data=df, x='Class', hue='satisfaction')
plt.title('Class vs Satisfaction')
plt.show()
```

# **TRIVARIATE**

```python
pd.crosstab(
    [df['Class'], df['Customer Type']],
    df['satisfaction']
)
```

# **CORRELATION**

```python
sns.heatmap(
    df.select_dtypes(include=['int64','float64']).corr(),
    cmap='YlGnBu'
)
plt.title('Correlation Heatmap')
plt.show()
```
